```python
dx = 5 # dimensao x do retangulo de selecao
dy = 5 # dimensao y do retangulo de selecao

px0 = 2 # deslocamento original que determina a posicao do retangulo na direcao x
py0 = 2 # deslocamento original que determina a posicao do retangulo na direcao x

px1 = 0 # deslocamentos secundarios para testar os calculos, eixo x
py1 = 1 # deslocamentos secundarios para testar os calculos, eixo y

px2 = 0 # deslocamentos secundarios para testar os calculos, eixo x
py2 = 2 # deslocamentos secundarios para testar os calculos, eixo y

px3 = 3 # deslocamentos secundarios para testar os calculos, eixo x
py3 = 3 # deslocamentos secundarios para testar os calculos, eixo y

ret0 = np.array([[0]*dx]*dy)
ret1 = np.array([[0]*dx]*dy)
ret2 = np.array([[0]*dx]*dy)
ret3 = np.array([[0]*dx]*dy)

for j in range (dy):
    for i in range (dx):
        ret0[j,i] = dado_v[j+px0,i+py0]
        ret1[j,i] = dado_v[j+px0+px1,i+px0+py1]
        ret2[j,i] = dado_v[j+px0+px2,i+px0+py2]
        ret3[j,i] = dado_v[j+px0+px3,i+px0+py3]

print ret0
print ret1
print ret2
print ret3


plt.plot(ret0,ret1,'+b')
plt.plot(ret0,ret0,'-r')
plt.show()

plt.plot(ret0,ret2,'+b')
plt.plot(ret0,ret0,'-r')
plt.show()

plt.plot(ret0,ret3,'+b')
plt.plot(ret0,ret0,'-r')
plt.show()
```
